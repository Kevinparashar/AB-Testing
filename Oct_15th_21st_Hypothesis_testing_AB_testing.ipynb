```python
Statistical Analysis

Population: The entire data or entire possible observations.
    
Sample: A subset of observations taken from population. 
    As the sample size increases sample will represent the 
    population more closely(Law of Large Numbers).
    
Parameters: It's the property of population which 
    we are interested in and never know the exact 
    value unless we do analysis on entire population
    (which is never the case) eg. mean( μ).
    
Estimates: It's sample idea/value about the population parameters. 
    The entire goal of statistics is to make these sample estimates 
    as close as population parameters eg, average( x¯) 
    is the best possible sample estimate of  μ .
    
Descriptive Statistics: It's for summarizing data.
    
Inferential Statistics: It's for drawing conclusions 
    about the population from samples eg., 
    estimating population mean( μ ) using sample average( x¯ ).
    
Parametric Statistics: Statistical methods where we assume a 
    distribution of the data such as Gaussian.
    
Non-Parametric Statistics: Statistical methods where we do not assume 
    any distribution of the data ie., distribution free.
    
Statistical Hypothesis Tests: Methods that quantify the likelihood of 
    observing the result given an assumption or expectation about the result. 
    We will talk on this more later.
    
Estimation Statistics: Methods that quantify the uncertainty of a result 
    using confidence intervals.
```

```python
#sports training example

Suppose we want to know whether or not a certain training program is able to 
increase the max vertical jump (in inches) of college basketball players.
To test this, we may recruit a simple random sample of 20 college basketball 
players and measure each of their max vertical jumps. Then, we may have each 
player use the training program for one month and then measure their max vertical 
jump again at the end of the month.


A paired samples t-test is commonly used in two scenarios:

    1. A measurement is taken on a player before and after training program – 
    e.g. the max vertical jump of college basketball players is measured before 
    and after participating in a training program.
    
        
In both cases we are interested in comparing the mean measurement between two groups 
in which each observation in one sample can be paired with an observation 
in the other sample.
```

```python
before_training=[22,20,19,24,25,25,28,22,30,27,24,18,16,19,19,28,24,25,25,23]
after_training=[24,22,23,27,28,26,28,24,30,29,25,20,17,18,18,28,26,27,27,24]
```

```python
The null hypothesis --- nothing is extra ordinary .. every thing or situation is as usual

Ho= the two population mean are equal.
H1= the two population means are not equal
```

```python
import scipy.stats as stats
```

```python
stats.ttest_rel(before_training,after_training)
```

```python
p<0.05--- we are able to reject the  null hypothesis
```

```python
variant_group=df[df["VARIANT_NAME"]=="variant"].drop_duplicates()
control_group=df[df["VARIANT_NAME"]=="control"].drop_duplicates()
```

```python
# we can notice an extreme increase in the Revenue from 99 percentile to 100 percentile.
#to ensure a better result in any of statistical test , we should take care the outliers.
```

```python
conf_interval=sms.DescrStatsW(df["REVENUE"]).tconfint_mean()
```

```python
conf_interval
```

```python
Q1=df["REVENUE"].quantile(0.01)
Q3=df["REVENUE"].quantile(0.99)
IQR=Q3-Q1
up_limit=Q3+1.5*IQR
low_limit=Q1-1.5*IQR
```

```python
print(up_limit,low_limit)
```

```python
# capping approach of outlier treatment  
```

```python
df["REVENUE"]=np.where(df["REVENUE"]>up_limit,up_limit,np.where(df["REVENUE"]<low_limit,low_limit,df["REVENUE"]))
```

```python
conf_interval=sms.DescrStatsW(df["REVENUE"]).tconfint_mean()
conf_interval
```

```python
# IT IS WITH 95% CONFIDENCE LEVEL THE REVENUE IS 0.04494625645609369, 0.06572259517632796 (Confidence interval)
```

```python
df.describe([0,0.05,0.10,0.95,0.99,1]).T
```

```python
# AB test --- (hypothesis testing) aim is to compare the revenue of two variants A and B and determine 
# if there is a statistically  significant different between them
```

```python
df.groupby("VARIANT_NAME").agg({"REVENUE":"mean"})
```

```python
# based on the above result, it is clear that revenue generated by control grop is better the variant group.
# but we can not conclude this statistically signifiant.
#this different can be a merely a result of random chance.
```

```python
# conducting AB testing (Hypothesis testing)
```

```python
Ho=no different the revenue of both group contol of variant are same.
H1=there is statistically significant difference in two group revenue .
```

```python
# lets check our data is normally distributed or not
```

```python
# sapiro test is also use to confirm whether our data is normally distributed or not 
Ho=data is notmally distribued
H1= not nomally distributed
```

```python
test_stat,p_value=shapiro(df.loc[df["VARIANT_NAME"]=="variant","REVENUE"])
print("Variant group",test_stat,p_value)
test_stat,p_value=shapiro(df.loc[df["VARIANT_NAME"]=="control","REVENUE"])
print("control group",test_stat,p_value)
```

```python
# we are able to reject the null hypothesis that data is normally distributed.
# means out data is not nomally distributed.
```

```python
# Mann-Whitney U 
```

```python
test_stat,p_value=mannwhitneyu(df.loc[df["VARIANT_NAME"]=="variant","REVENUE"],
                              df.loc[df["VARIANT_NAME"]=="control","REVENUE"])

print("Test stats:",test_stat)
print("p value: ",p_value)
```

```python
p>0.05 we fail to reject the null hypothesis.
conclusion:-no different the revenue of both group contol of variant are same.
```
